```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('cleaned_dataset.csv')
df.head()
```

```python
df['Family Income'] = df['Family Income'].fillna(df['Family Income'].median())
df['Family Size'] = pd.to_numeric(df['Family Size'], errors='coerce')
df['Family Size'] = df['Family Size'].fillna(df['Family Size'].median())
```

```python
df['House Size (Marlas)'] = df['House Size (Marlas)'] * 250
df.rename(columns={'House Size (Marlas)': 'House Size (Sqft)'}, inplace=True)
df['Personal Space (Sqft)'] = df['House Size (Sqft)'] / df['Family Size']
df.to_csv('Final_Dataset.csv')
```

```python
print(df.dtypes)
df.describe()
```

```python
numerical_cols = [
    'Higher Secondary Marks Percentage (%)',
    'NET Marks',
    'Family Size',
    'Personal Space (Sqft)'
]

for column in numerical_cols:
    if column != 'SGPA':
        plt.figure()
        plt.scatter(df[column], df['SGPA'])
        plt.xlabel(column)
        plt.ylabel('SGPA')
        plt.title(f'{column} vs SGPA')
        plt.grid(True)
        plt.show()
```

```python
ordinal_cols = [
    'Parents Education Level',
    'Family Income',
    'House Size (Sqft)',
    'Family Size',
    'Stress Level',
    'Confidence Level',
    'Study Consistency Level',
    'Participation in Extra-Curricular Activities',
    'Available Emotional Support',
    'Sleeping Level',
    'Attendance Level',
    'Access to Resources',
    'Understanding of Subject'
]

# Make sure SGPA is numeric
df['SGPA'] = pd.to_numeric(df['SGPA'], errors='coerce')

# Create violin plots
for col in ordinal_cols:
    plt.figure(figsize=(8, 5))
    sns.violinplot(x=col, y='SGPA', data=df)
    plt.title(f'{col} vs SGPA')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
